```python
import pandas as pd
import numpy as np
from matplotlib import pyplot as plt
from sklearn.linear_model import LinearRegression
```

```python
df = pd.read_csv('../../data/raw/regression_example.csv')

df.head()
```

```python
model = LinearRegression()

model.fit(df.X.values.reshape(-1,1), df.y.values.reshape(-1,1))
```

```python
y_hat2 = model.predict(df.X.values.reshape(-1,1))
erro_quadratico_medio2 = np.mean((y_hat2 - df.y.values.reshape(-1,1))**2)

plt.scatter(df.X, df.y)
plt.plot(df.X.values.reshape(-1,1), y_hat2, color='k')
plt.title(f"Erro = {erro_quadratico_medio2:.5f}")
plt.show()
```
