# Repetição (loops)
 - Dois tipos: `while`, `for`

## while
 - `break`
 - `continue`
 - `else`

```python
numero = 17
inicio = 2
corrente = inicio
while corrente < numero - 1:
    if numero % corrente == 0:
        break
    corrente += 1
else:
    print('o número é primo')
```

## for
 - `break`
 - `continue`
 - `range()`
 - `else`
 - `pass` e `...`

```python
list(range(1, 100, 2))
```

```python
help(range)
```

```python
for i in range(0, 101):
    if i % 2 == 0:
        print(i)
        continue
```
