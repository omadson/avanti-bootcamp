# Seleção (estruturas condicionais)
### `if`

```python
foo = 8

if foo >= 7:
    print('passou!')
```

### `if..else`

```python
foo = 32
if foo > 10:
    print('foo é maior que 10')
else:
    print('foo é menor que 10')
```

### `if..elif..else`

```python
numero = 90

if (numero % 3 == 0) and (numero % 5 == 0):
    print('FizzBuzz')
elif numero % 3 == 0:
    print('Fizz')
elif numero % 5 == 0:
    print('Buzz')
else:
    print(numero)
```

### short hand if

```python
numero = 6
if numero % 3 == 0: print('Fizz')
```

### short hand if..else (operador ternário)

```python

# if foo > 10:
#     print('foo é maior que 10')
# else:
#     print('foo é menor que 10')

foo = 32

print('foo é maior que 10' if foo > 10 else 'foo é menor que 10')
```

```python
numero = 3
print("Fizz" if numero % 3 == 0 else "Buzz")
```
