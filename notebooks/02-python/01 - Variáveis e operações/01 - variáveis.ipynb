# Variáveis

 - Containers para armazenar valores
 - Sem declaração de variáveis
 - Criada no momento da atribuição

```python
foo = 42 # int
bar = "Python" # str
spam = 1.4 # float
```

## Casting (conversão explícita)

```python
baz = str(foo)
```

## Identificadores
 - possíveis nomes
 - camel case `minhaVariável`
 - pascal case `MinhaVariável`
 - Snake case `minha_variável`

```python
my_variable = 2
```

## Valores para variáveis
 - multiplos para multiplos
 - único para multiplos

```python
foo = bar = spam = 42
```

## Built-in Data Types
 - texto: `str`
 - números: `int`, `float`, `complex`
 - sequências: `list`, `tuple`, `range`
 - chave-valor: `dict`
 - conjuntos: `set`, `frozenset`
 - booleanos: `bool`
 - tipos binários: `bytes`, `bytearray`, `memoryview`

```python
foo = True
bar = False

type(bar)
```
