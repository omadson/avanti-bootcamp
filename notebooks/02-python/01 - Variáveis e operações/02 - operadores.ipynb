# Operadores

 - Executar operações com variáveis e valores
 
 ### Matemáticos e de atribuição
 
 - Adição: `+`
 - Subtração: `-`
 - Multiplicação: `*`
 - Divisão: `/`
 - Módulo: `%`
 - Potenciação: `**`
 - Divisão inteira: `//`
 - Atribuição: `=`, `+=`, `-=`, `*=`, ...

```python
foo = 10/3
print(type(foo))
```

### Relacionais
 - Igualdade `==`
 - Desigualdade `!=`
 - Maior que `>`
 - Menor que `<`
 - Maior ou igual`>=`
 - Maior ou igual`<=`

```python
foo = 10
bar = 5
spam = 3

foo > bar < spam
```

### Lógicos
 - `and`
 - `or`
 - `not`

```python
foo > bar and bar < spam
```

### Identidade e associação
 - `is`
 - `in`

```python
'Ciência' in 'Ciência de dados'
```
