# Tuplas
 - Utilizadas para armazenar multiplos itens em uma mesma variável
 - Ordenadas, imutáveis e permitem valores duplicados
 - Podem ser escritas entre parênteses

```python
tupla_frutas = ('manga', 'jaca', 'pera', 'goiaba', 'maçã', 'jaca', 'uva', 'jaca')
nova_tupla_frutas = ('cajá',)
notas = (10, 7, 9.5, 7)
tupla_diversa = (42, notas, 'Python', 1.5)
tupla_unitaria = (73,)
```

```python
notas = list(notas)
tupla_diversa = (42, notas, 'Python', 1.5)
```

## Itens de uma tupla
 - Itens são acessados usando a sintaxe: `<tupla>[<pocisão>]`
   - O primeiro item tem índice `[0]`, o segundo item tem índice `[1]` etc.
   - O ultimo item tem índice `[-1]`, o penultimo item tem índice `[-2]` etc.
   - Intervalos de itens (slices): `[<inicio>:<fim>:<passo>]`
 - Checar se um valor existe: `<valor> in <tupla>`
 - Percorrer itens de uma tupla usando `for`

```python
for fruta in tupla_frutas:
    print(fruta)
```

## Operações com tuplas
 - Unindo (concatenando) duas tuplas
 - Repedindo uma tupla

```python
print(tupla_frutas * 6)
```

## Funções e tuplas
 - Conversão explícita (construtor): `tuple(<valor>)`
 - Tamanho: `len(<tupla>)`
 - Soma: `sum(<tupla>)`
 - Maior valor: `max(<tupla>)`
 - Menor valor: `min(<tupla>)`
 - Ordenar valores: `sorted(<tupla>)`

## Métodos de tuplas
 - Encontrar valores: `<tupla>.index(<valor>)`
 - Contar valores: `<tupla>.count(<valor>)`

```python
help(tupla_frutas.count)
```
