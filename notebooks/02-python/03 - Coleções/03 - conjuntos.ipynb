# Conjuntos
 - Utilizadas para armazenar multiplos itens em uma mesma variável
 - Desordenados e não indexável
 - Escritos usando chaves
 - Imutáveis

## Itens de um conjunto
 - Não tem uma ordem definida
 - Podem aparecer em uma ordem diferente sempre que você os usa e não podem ser referenciados por índice ou chave
 - Não podemos alterar os itens após a criação do conjunto
 - Não podem ter dois ítens iguais

## Operações com conjuntos
 - União: `|`
 - Interceção: `&`
 - Diferença: `-`

## Funções e conjuntos
 - Conversão explícita (construtor): `set(<valor>)`
 - Soma: `sum(<conjunto>)`
 - Maior valor: `max(<conjunto>)`
 - Menor valor: `min(<conjunto>)`
 - Valores ordenados: `sorted(<conjunto>)`

## Métodos de conjuntos
 - Adicionar elemento: `<conjunto>.add(<valor>)`
 - Adicionar vários elementos: `<conjunto>.update(<conjunto>)`
 - Verifica se é subconjunto: `<conjunto>.issubset(<conjunto>)`
 - Verifica se é superconjunto: `<conjunto>.issuperset(<conjunto>)`
 - Remover elementos por valor: `<conjunto>.remove(<valor>)`
