# Strings
 - As strings em python são colocadas entre aspas simples ou aspas duplas.
 - Também podemos criar strings com múltiplas linhas
 - A função `print(...)` serve para imprimir uma string
 - São imutáveis

```python
spam = 'Python'
print(spam)
```

## Itens de uma string
 - Bastante parecidos com ítens em tuplas

```python
spam[1]
```

## Operações com strings
 - Concatenação: `+`
 - Multiplicação: `*`

```python
print(spam*20)
```

## Funções e strings
 - Conversão explícita (construtor): `str(<valor>)`
 - Tamanho: `len(<string>)`

## Métodos de strings (alguns)
 - Deixar a primeira maiúscula: `<string>.capitalize()`
 - Contar ocorrências: `<string>.count(<valor>)`
 - Checa se termina com um valor: `<string>.endswith(<string>)`
 - Checa se começa com um valor: `<string>.startswith(<string>)`
 - Converte para maiúsculas: `<string>.upper()`
 - Converte para minúsculas: `<string>.lower()`
 - Pesquisa por uma substring: `<string>.find(<string>)`
 - Quebra string: `<string>.split(<string>)`
