# Dicionários
 - Utilizadas para armazenar informação em pares chave valor
 - Ordenados (Python 3.6+), mutáveis e não permitem duplicatas
 - Escritos utilizando chaves e seus itens são pares chave e valor separados por dois pontos
 - Cada item é separado por uma vírgula

```python
nome_completo = {'nome': 'Rodolfo', 'sobrenome': 'Ricardo'}
nome_completo2 = dict(nome='Lucas', sobrenome='Braga')
print(nome_completo['sobrenome'])
```

## Itens de um dicionário
 - Podem ser de qualquer tipo (tanto a chave quanto o valor)
 - Acesso do valor através da chave
 - Checar se uma chave existe: `<chave> in <dicionário>`
 - Mudar valor: `<dicionario>[<chave>] = <novo_valor>` (a `<chave>` já existe)
 - Adicionar um novo ítem: `<dicionario>[<chave>] = <novo_valor>` (a `<chave>` não existe)

```python
print(nome_completo2)
```

## Funções e dicionários
 - Conversão explícita (construtor): `dict(<lista_ou_tupla_de_chave_valor>)`
 - tamanho: `len(<dicionario>)`

## Métodos de dicionários
 - Retornar o valor dada uma chave: `<dicionário>.get(<chave>)`
 - Modificar o valor, data uma chave e o novo valor: `<dicionário>.set(<chave>, <novo_valor>)`
 - Retornar uma lista dos itens: `<dicionário>.items()`
 - Retorna uma lista com os valores: `<dicionário>.values()`
 - Retorna uma lista com as chaves: `<dicionário>.keys()`
 - Remover um item dada uma chave:  `<dicionário>.pop(<chave>)`
 - Copiar um dicionário:  `<dicionário>.copy()`
 - Remover um item dada uma chave:  `<dicionário>.pop(<chave>)`
