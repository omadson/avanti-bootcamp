# Listas
 - Utilizadas para armazenar multiplos itens em uma mesma variável
 - Ordenadas, mutáveis e permitem valores duplicados
 - Podem ser escritas usando colchetes

```python
# criando uma tupla
frutas = ["laranja", "maçã", "pera", "kiwi", "maçã"]
moveis = ["mesa", "cadeira", "armário", "roupeiro", "sofá", "cama", "rack", "poltrona"]
notas = [10, 7.5, 9, 5.1, 8.3]
resultado_enquete = [True, False, True, True, False, False, True]
# criando uma tupla vazia
lista_vazia = []
# criando uma tupla unitária
lista_unitaria = ["alguma coisa"]
outra_lista_unitaria = [42,]
# criando uma tupla diversa
lista_diversa = [42, (), "Python", frutas, True]
```

## Itens de uma lista
 - Itens são acessados usando a sintaxe: `<lista>[<pocisão>]`
   - O primeiro item tem índice `[0]`, o segundo item tem índice `[1]` etc.
   - O ultimo item tem índice `[-1]`, o penultimo item tem índice `[-2]` etc.
   - Intervalos de itens (slices): `[<inicio>:<fim>:<passo>]`
 - Itens podem ser modificados usando o operador de atribuição `=`
 - Checar se um valor existe: `<valor> in <lista>`
 - Percorrer itens de uma lista usando `for`

```python
print(frutas)
frutas[4] = 'cajá'
print(frutas)
```

```python
print(frutas)
```

```python
# imprimindo o terceiro valor de uma lista
print("O segundo ítem da tupla de frutas é", frutas[2])
```

```python
# imprimindo o penultimo valor de uma lista
print("O penultimo ítem da tupla de móveis é", moveis[-2])
```

```python
# imprimindo do segundo ao quinto item de uma lista
print("Do segundo ao penultimo", moveis[1:5])
```

```python
# imprimindo intervalos
print("Imprimindo até o quinto elemento", moveis[:5])
print("Imprimindo até o penultimo elemento", moveis[:-1])
print("Imprimindo todos os elementos", moveis[:])
print("Imprimindo todos os elementos", moveis[::])
print("Imprimindo na ordem inversa", moveis[::-1])
```

```python
# mudando o segundo valor de uma lista 
frutas[1] = "jaca"
print(frutas)

# mudando vários valores de uma lista
moveis[:3] = [10]
print(moveis)
```

```python
# checando se os valores existem
if "jaca" in frutas:
    print("Tem goiaba na lista de frutas")
else:
    print("Não tem goiaba na lista de frutas")
```

```python
# percorrendo itens
for movel in moveis:
    print(movel)
```

## Operações com listas
 - Unindo (concatenando) duas ou mais listas
 - Repedindo uma lista

```python
tupla_unida = moveis + frutas
print('Unindo a lista de móveis e a lista de frutas', tupla_unida)
```

```python
tupla_repetida = moveis * 3
print('Repetindo a lista', tupla_repetida)
```

## Funções e listas
 - Conversão explícita (construtor): `list(<valor>)`
 - Tamanho: `len(<lista>)`
 - Soma: `sum(<lista>)`
 - Maior valor: `max(<lista>)`
 - Menor valor: `min(<lista>)`

```python
print("O tamanho da lista de móveis é", len(moveis))
print("A soma lista de notas é", sum(notas))
print("O maior valor da lista de notas é", max(notas))
print("O menor valor da lista de frutas é", min(frutas))
```

## Métodos de listas
 - Encontrar valores: `<lista>.index(<valor>)`
 - Contar valores: `<lista>.count(<valor>)`
 - Adicionar elemento ao final de uma lista`<lista>.append(<valor>)`
 - Copiar uma lista: `<lista>.copy()`
 - Inserir valores: `<lista>.insert(<pocisão>, <valor>)`
 - Remover elementos por posição: `<lista>.pop(<pocisão>)`
 - Remover elementos por valor: `<lista>.remove(<valor>)`

```python
frutas = ["laranja", "maçã", "pera", "kiwi", "maçã"]
print(frutas)
outra_lista = frutas.copy()
print(outra_lista)
outra_lista[1] = 'uva'
print(outra_lista)
print(frutas)
frutas.pop()
print(frutas)
frutas.remove('maçã')
print(frutas)
```

```python
frutas.append('abacate')
frutas.append('jaboticaba')
```

```python
print(frutas)
frutas.insert(frutas.index('abacate'), 'açaí')
print(frutas)
```
