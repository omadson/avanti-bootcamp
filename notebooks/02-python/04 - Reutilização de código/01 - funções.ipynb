# Funções

 - Bloco de código que executa quando é chamado
 - Podemos passar dados por parâmetros
 - Pode retornar um valor

## Definição

```python
def fizz_buzz(numero):
    if (numero % 3 == 0) and (numero % 5 == 0):
        print('FizzBuzz')
    elif numero % 3 == 0:
        print('Fizz')
    elif numero % 5 == 0:
        print('Buzz')
    else:
        print(numero)
```

## Chamada

```python
def imprimir_nome_sobrenome(nome, sobrenome='da Siva'):
    print(f'Nome: {nome}')
    print(f'Sobrenome: {sobrenome}')
```

```python
imprimir_nome_sobrenome(nome='Paulo')
```

```python
def retorna_h(a, b):
    return (a**2 + b**2)**(1/2)
```

```python
h = retorna_h(3, 4)

print(h)
```

# Argumentos
 - Valores padrão
 - Argumentos arbitrários: `*args`
 - Argumentos de Palavras-Chave: `**kwargs`

## Lambda functions


```python
retorna_h = lambda a, b: (a**2 + b**2)**(1/2)

h = retorna_h(3, 4)

print(h)
```
