## Criação/Leitura de um conjunto de dados (tabular)
 - Criação: `pd.DataFrame({<dicionario>,<lista>,<tupla>)`
 - Leitura: `pd.read_{csv,excel,json,parquet,picle,...}(<endereço_do_arquivo>)`

```python
df_gatos = pd.DataFrame({
    'nome': ['Fisher', 'Bayes'],
    'idade': [4, 2],
    'cor': ['branco e marrom', 'branco e laranja'],
    'peso': [5.5, 3.9]
})

# mesmo retorno
df_gatos = pd.DataFrame([
    {'nome': 'Fisher', 'idade': 4, 'cor': 'branco e marrom', 'peso': 5.5},
    {'nome': 'Bayes', 'idade': 2, 'cor': 'branco e laranja', 'peso': 4.9}
])

df = pd.read_csv('data/penguins.csv')

URL = 'https://raw.githubusercontent.com/mwaskom/seaborn-data/master/tips.csv'
df_tips = pd.read_csv(URL)
```

## Seleção de linhas
 - Primeiras linhas: `head(<numero_de_amostras>)`

 - Linhas aleatórios: `sample(<numero_de_amostras>)`
 - Slicing: `loc[], iloc[]`

## Aplicação de funções
 - Tablewise: `pipe()`
 - Row-wise ou Column-wise: `apply()`
 - Aggregação: `agg()` e `transform()`

## Métodos importantes
 - Queries: `query(<query>)`
 - Ordenação: `sort_index(), sort_values()`
 - Maiores e menores: `nlargest(), nsmallest()`
 - Selecionando por tipo: `select_dtypes()`
 - Agrupar: `group_by()`
 - Criar colunas: `assign()`
 - Deletar colunas: `drop()`
